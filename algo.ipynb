```python
import time
import matplotlib.pyplot as plt

def fibonacci(n):
    a, b = 0, 1
    sequence = []
    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b
    return sequence

# Input sizes
input_sizes = [5, 10, 20, 30, 40, 100, 500, 1000]
times = []

for n in input_sizes:
    start = time.time()
    fibonacci(n)
    end = time.time()
    times.append(end - start)

# Print execution results with complexities
print("===== Fibonacci Iterative Results =====")
for n, t in zip(input_sizes, times):
    print(f"Fibonacci({n}) -> Time Taken: {t:.8f} sec | "
          f"Time Complexity: O(n) | Space Complexity: O(n)")

# Plot results
plt.plot(input_sizes, times, marker='o', label="Execution Time")
for i, txt in enumerate(times):
    plt.text(input_sizes[i], times[i], f"{input_sizes[i]}", fontsize=9, ha='right')

plt.xlabel("Input size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Fibonacci Iterative Execution Time with Complexity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import time
import matplotlib.pyplot as plt

# Naive recursive Fibonacci
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

def test_fib_recursive(max_n=30):
    times = []
    inputs = list(range(1, max_n+1))
    for n in inputs:
        start = time.time()
        fib_recursive(n)
        end = time.time()
        times.append(end - start)
    return inputs, times

# Plot
def plot_fib_recursive():
    inputs, times = test_fib_recursive()
    plt.plot(inputs, times, marker="o", label="Recursive Fibonacci")
    plt.xlabel("Input n")
    plt.ylabel("Execution Time (s)")
    plt.title("Naive Recursive Fibonacci Time Complexity")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_fib_recursive()
```

```python
import time
import matplotlib.pyplot as plt

# DP Fibonacci
def fib_dp(n):
    if n <= 1:
        return n
    dp = [0, 1]
    for i in range(2, n+1):
        dp.append(dp[i-1] + dp[i-2])
    return dp[n]

def test_fib_dp(max_n=5000):
    times = []
    inputs = list(range(1, max_n+1, 200))
    for n in inputs:
        start = time.time()
        fib_dp(n)
        end = time.time()
        times.append(end - start)
    return inputs, times

def plot_fib_dp():
    inputs, times = test_fib_dp()
    plt.plot(inputs, times, marker="o", color="green", label="DP Fibonacci")
    plt.xlabel("Input n")
    plt.ylabel("Execution Time (s)")
    plt.title("Dynamic Programming Fibonacci")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_fib_dp()
```

```python
import time
import matplotlib.pyplot as plt
import random

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


def test_merge_sort():
    sizes = [100, 500, 1000, 2000, 5000]
    times = []

    for size in sizes:
        arr = random.sample(range(size * 10), size)
        start = time.time()
        merge_sort(arr)
        end = time.time()
        times.append(end - start)
        print(f"Input size: {size}, Time taken: {end - start:.6f} sec")

    # Plot graph
    plt.plot(sizes, times, marker='o', label="Merge Sort")
    plt.xlabel("Input Size (n)")
    plt.ylabel("Time (seconds)")
    plt.title("Merge Sort Time Complexity")
    plt.legend()
    plt.show()

test_merge_sort()
```

```python
import time
import matplotlib.pyplot as plt
import random

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[random.randint(0, len(arr)-1)]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

sizes = [100, 1000, 5000, 10000]
times = []
for n in sizes:
    arr = [random.randint(0, 10000) for _ in range(n)]
    start = time.perf_counter()
    quick_sort(arr)
    end = time.perf_counter()
    times.append(end - start)

plt.plot(sizes, times, marker="o")
plt.xlabel("Array Size")
plt.ylabel("Time (s)")
plt.title("Quick Sort: Time vs Input Size")
plt.show()
```

```python
import time
import matplotlib.pyplot as plt
import random

def insertion_sort(arr):
    a = arr[:]
    for i in range(1, len(a)):
        key = a[i]
        j = i - 1
        while j >= 0 and a[j] > key:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = key
    return a

sizes = [100, 300, 600, 1000]
times = []
for n in sizes:
    arr = [random.randint(0, 10000) for _ in range(n)]
    start = time.perf_counter()
    insertion_sort(arr)
    end = time.perf_counter()
    times.append(end - start)

plt.plot(sizes, times, marker="o")
plt.xlabel("Array Size")
plt.ylabel("Time (s)")
plt.title("Insertion Sort: Time vs Input Size")
plt.show()
```

```python
import time
import matplotlib.pyplot as plt
import random

def bubble_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
                swapped = True
        if not swapped:
            break
    return a

sizes = [100, 300, 600, 1000]
times = []
for n in sizes:
    arr = [random.randint(0, 10000) for _ in range(n)]
    start = time.perf_counter()
    bubble_sort(arr)
    end = time.perf_counter()
    times.append(end - start)

plt.plot(sizes, times, marker="o")
plt.xlabel("Array Size")
plt.ylabel("Time (s)")
plt.title("Bubble Sort: Time vs Input Size")
plt.show()
```

```python
import time
import random
import matplotlib.pyplot as plt
from memory_profiler import memory_usage

# ---------------------------
# Merge Sort Implementation
# ---------------------------
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr


# ---------------------------
# Experimental Profiling
# ---------------------------
def run_experiment():
    input_sizes = [100, 500, 1000, 2000, 5000]
    times = []
    mem_usage = []

    for size in input_sizes:
        arr = [random.randint(1, 10000) for _ in range(size)]

        start_time = time.time()
        mem_before = memory_usage()[0]

        merge_sort(arr)

        mem_after = memory_usage()[0]
        end_time = time.time()

        times.append(end_time - start_time)
        mem_usage.append(mem_after - mem_before)

    # ---------------------------
    # Plot Graph
    # ---------------------------
    plt.plot(input_sizes, times, marker='o')
    plt.title("Merge Sort Execution Time")
    plt.xlabel("Input Size (n)")
    plt.ylabel("Time (seconds)")
    plt.grid(True)
    plt.show()

    return times, mem_usage


if __name__ == "__main__":
    times, mem_usage = run_experiment()
    print("Execution Times:", times)
    print("Memory Usage (MB):", mem_usage)
```
